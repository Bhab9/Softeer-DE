```python
combination = df.groupby(['gear', 'am']).size().reset_index(name='count')
combination
```

```python
combination2 = pd.crosstab(df['gear'], df['am'])
combination2
```

```python
import matplotlib.pyplot as plt
```

```python
x_labels = [f"({g}, {a})" for g, a in zip(combination['gear'], combination['am'])]
plt.figure()
plt.bar(x_labels, combination['count'])
plt.title('# of Cars by Gear x Transmission')
plt.xlabel('(Gear, Auto Transmission)')
plt.ylabel('# of Cars')
plt.show()
```

```python
combination2.plot(kind='bar')
plt.title('# of Cars by Gear x Transmission')
plt.xlabel('(Gear, Auto Transmission)')
plt.ylabel('# of Cars')
plt.show()
```

```python
numeric_cols = df.select_dtypes(include='number').columns

n_cols = 3
n_rows = 4
plt.figure(figsize=(5*n_cols, 4*n_rows))

for i, col in enumerate(numeric_cols, 1):
    plt.subplot(n_rows, n_cols, i)
    plt.hist(df[col], bins=10, edgecolor='black')
    plt.title(col)
    plt.grid(True)
    plt.tight_layout()

plt.suptitle("Histograms of Variables", fontsize=20)
plt.subplots_adjust(top=0.93)
plt.show()
```

```python
plt.figure()
plt.scatter(df['cyl'], df['hp'])
plt.title("Cylinder vs Horse Power")
plt.xlabel('Cylinder')
plt.ylabel('Horse Power')
plt.show()
```

```python
plt.figure()
plt.scatter(df['mpg'], df['hp'])
plt.title("MPG vs Horse Power")
plt.xlabel('MPG')
plt.ylabel('Horse Power')
plt.show()
```
