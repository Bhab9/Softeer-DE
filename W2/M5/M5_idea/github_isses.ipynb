````python
import pandas as pd
import re
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
````

````python
df = pd.read_csv("github_issues.csv")
df.head()
````

````python
df["body"].head(10)
````

````python
def clean_text(text):
    text = text.lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"[^a-z\s]", " ", text)
    text = re.sub(r"\s+", " ", text)
    return text.strip()
````

````python
urgent_words = [
    "urgent", "asap", "immediately", "blocker", "critical",
    "cannot", "can't", "failed", "broken", "error"
]

rework_words = [
    "again", "still", "retry", "reopen", "not working",
    "doesn't work", "fail", "issue persists"
]

conflict_words = [
    "not my", "as mentioned", "already said",
    "please read", "i explained", "expected behavior"
]
````

````python
def count_keywords(text, keywords):
    return sum(text.count(k) for k in keywords)

df["urgent_score"] = df["issue_title"].apply(lambda x: count_keywords(x, urgent_words))
df["rework_score"] = df["issue_title"].apply(lambda x: count_keywords(x, rework_words))
df["conflict_score"] = df["issue_title"].apply(lambda x: count_keywords(x, conflict_words))
````

````python
analyzer = SentimentIntensityAnalyzer()
df["sentiment"] = df["issue_title"].apply(lambda x: analyzer.polarity_scores(x)["compound"])
````

````python
df["risk_score"] = (
    df["urgent_score"] * 2
    + df["rework_score"] * 1.5
    + df["conflict_score"] * 2
    + (-df["sentiment"])  # 부정일수록 점수 ↑
)
````

````python
df.sort_values("risk_score", ascending=False)[
    ["issue_url", "issue_title", "risk_score"]
].head(10)
````

````python
df["repo"] = df["issue_url"].apply(lambda x: x.split("/")[4])

repo_risk = df.groupby("repo")["risk_score"].mean().sort_values(ascending=False)
````
